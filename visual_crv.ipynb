```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation
import random
```

```python
mode = "lstm" #vlstm or social
testset = 0

if mode == "social":
    file_dir = "log_social_model/" + str(testset) +"/"
    jpg_title = "social"
else:
    file_dir = "log_vlstm/" + str(testset) +"/"
    jpg_title = "vlstm"
```

```python
val_crv_data = []
with open(file_dir+"log_curve_v.txt","rt") as data:
    
    lines = data.readlines()
    for line in lines:
        
        lineData = line.strip().split(',')
        val_crv_data.append(lineData)
data.close()
```

```python
V_Crvs = []
for i in range(len(val_crv_data)-1):
    
    V_Crvs.append(float(val_crv_data[i+1][1]))
```

```python
test_crv_data = []
with open(file_dir+"log_curve_t.txt","rt") as data:
    
    lines = data.readlines()
    for line in lines:
        
        lineData = line.strip().split(',')
        test_crv_data.append(lineData)
data.close()
```

```python
T_Crvs = []
for i in range(len(test_crv_data)):
    
    T_Crvs.append(float(test_crv_data[i][1]))
```

```python
# initialize figure
fig, ax = plt.subplots()
ax.set_xlim(0, 55)
ax.set_ylim(-5, 0)
plt.plot(V_Crvs, label = "validation crv")
plt.plot(T_Crvs, label = "training crv")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.title("Testdataset" + ": " +str(testset))
plt.savefig("crv_plots/"+jpg_title+str(testset)+".jpg")
plt.show()
```
